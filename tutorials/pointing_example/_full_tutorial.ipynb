# Rex-Omni Pointing Tutorial

This notebook demostrates how to use Rex-Omni for pointing tasks.


## Step 1: Initialize Rex Omni Model

```python
# Import required libraries
import torch
from PIL import Image
from rex_omni import RexOmniWrapper, RexOmniVisualize
import matplotlib.pyplot as plt
import numpy as np

# Rex-Omni supports both Transformers and VLLM backends by switching the backend parameter.
model_path = "IDEA-Research/Rex-Omni"  # Replace with your model path

print("ðŸš€ Initializing Rex Omni model...")

rex_model = RexOmniWrapper(
    model_path=model_path,
    backend="transformers",  # Choose "transformers" or "vllm"
    max_tokens=2048,
    temperature=0.0,
    top_p=0.05,
    top_k=1,
    repetition_penalty=1.05,
)
```

## Example 1: Object Pointing

```python
image_path = "test_images/boxes.jpg"  # Replace with your image path
image = Image.open(image_path).convert("RGB")
print(f"âœ… Image loaded successfully!")
print(f"ðŸ“ Image size: {image.size}")

# detect in word level
# Define categories to detect
categories = [
    "open boxes",
    "closed boxes"
]
# inference
results = rex_model.inference(images=image, task="pointing", categories=categories)

result = results[0]
if result["success"]:
    predictions = result["extracted_predictions"]
    vis_image = RexOmniVisualize(
        image=image,
        predictions=predictions,
        font_size=30,
        draw_width=10,
        show_labels=True,
    )
    # display the visualization
    plt.figure(figsize=(10, 8))
    plt.imshow(vis_image)
    plt.axis("off")
    plt.title("Object Pointing Visualization")
    plt.show()

else:
    print(f"Inference failed: {result['error']}")
```
